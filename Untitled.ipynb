```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
TEAM_1_PLAYER = "/home/morote/Pictures/team1_black.png"
TEAM_2_PLAYER = "/home/morote/Pictures/team2_white.png"
TEAM_3_PLAYER = "/home/morote/Pictures/team2_white2.png"
TEAM_4_PLAYER = "/home/morote/Pictures/team1_black2.png"
```

```python
team1_image = cv2.imread(TEAM_1_PLAYER)
team2_image = cv2.imread(TEAM_2_PLAYER)
team3_image = cv2.imread(TEAM_3_PLAYER)
team4_image = cv2.imread(TEAM_4_PLAYER)
```

```python
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
def get_color_histogram(image):
    im = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    red, green, blue = cv2.split(im)

    hist_red = cv2.calcHist([red], [0], None, [256], [0, 256])
    hist_green = cv2.calcHist([green], [0], None, [256], [0, 256])
    hist_blue = cv2.calcHist([blue], [0], None, [256], [0, 256])
    return hist_red, hist_green, hist_blue
```

```python
players = [team1_image, team2_image, team3_image, team4_image]

for player in players:
    for player2 in players:
        hist1_red, hist1_green, hist1_blue = get_color_histogram(player)
        hist2_red, hist2_green, hist2_blue = get_color_histogram(player2)

        cv2.normalize(hist1_red, hist1_red, 0, 1, cv2.NORM_MINMAX)
        cv2.normalize(hist1_green, hist1_green, 0, 1, cv2.NORM_MINMAX)
        cv2.normalize(hist1_blue, hist1_blue, 0, 1, cv2.NORM_MINMAX)

        cv2.normalize(hist2_red, hist2_red, 0, 1, cv2.NORM_MINMAX)
        cv2.normalize(hist2_green, hist2_green, 0, 1, cv2.NORM_MINMAX)
        cv2.normalize(hist2_blue, hist2_blue, 0, 1, cv2.NORM_MINMAX)

        chi_square_distance_r = cv2.compareHist(hist1_red, hist2_red, cv2.HISTCMP_BHATTACHARYYA)
        chi_square_distance_g = cv2.compareHist(hist1_green, hist2_green, cv2.HISTCMP_BHATTACHARYYA)
        chi_square_distance_b = cv2.compareHist(hist1_blue, hist2_blue, cv2.HISTCMP_BHATTACHARYYA)

        print(f"Dist player1 to player2: {chi_square_distance_r}, {chi_square_distance_g}, {chi_square_distance_b}")
    print("-----------------------------------")
```
